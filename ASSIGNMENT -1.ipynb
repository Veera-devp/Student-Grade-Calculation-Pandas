```python
df= data.drop(labels=["Id","Name", "Gender"], axis=1)
```

```python
df['Subject1 Marks'] = pd.to_numeric(df['Subject1 Marks'], downcast='float', errors='coerce')
df['Subject2 Marks'] = pd.to_numeric(df['Subject2 Marks'],downcast='float', errors='coerce')
df['Subject3 Marks'] = pd.to_numeric(df['Subject3 Marks'],downcast='float', errors='coerce')
df['Subject4 Marks'] = pd.to_numeric(df['Subject4 Marks'],downcast='float', errors='coerce')
df['Subject5 Marks'] = pd.to_numeric(df['Subject5 Marks'],downcast='float', errors='coerce')
df['Subject6 Marks'] = pd.to_numeric(df['Subject6 Marks'],downcast='float', errors='coerce')
```

```python
df.isnull().any()
```

```python
df[['Subject2 Marks','Subject3 Marks','Subject4 Marks','Subject5 Marks','Subject6 Marks']] = df[['Subject2 Marks','Subject3 Marks','Subject4 Marks','Subject5 Marks','Subject6 Marks']].fillna(value=df[['Subject2 Marks','Subject3 Marks','Subject4 Marks','Subject5 Marks','Subject6 Marks']].mean())
df
```

```python
v1=d["Subject1 Marks"]
v2=d["Subject2 Marks"]
v3=d["Subject3 Marks"]
v4=d["Subject4 Marks"]
v5=d["Subject5 Marks"]
v6=d["Subject6 Marks"]
total=v1+v2+v3+v4+v5+v6
df_t=pd.DataFrame(total)
df_t=df_t.rename(columns={0:"Total marks"})
pd.options.display.float_format = '{: .1f}'.format
df_t
```

```python
l0=[]
for i in df_t['Total marks']:
    z=i/6
    l0.append(z)
l0
marks=pd.DataFrame(l0)
marks=marks.rename(columns={0:"Total marks"})
pd.options.display.float_format = '{: .0f}'.format
marks['Total marks']
marks['Total marks'] = marks['Total marks'].astype(int)
```

```python
l=[]
for z in marks['Total marks']:
    if(z>=80 and z<=100):
        l.append("O")
    elif(z>=70 and z<=79):
        l.append("A+")
    elif(z>=60 and z<=69):
        l.append("A")   
    elif(z>=55 and z<=59):
        l.append("B+")  
    elif(z>=50 and z<=54):
        l.append("B")  
    elif(z>=45 and z<=49):
        l.append("C")
    elif(z>=40 and z<=44):
        l.append("P")
    elif(z>=0 and z<=39):
        l.append("F")
cgpa=pd.DataFrame(l)
cgpa=cgpa.rename(columns={0:"CGPA"})
cgpa
```
